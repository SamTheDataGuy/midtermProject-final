```python
#Google Places API notebook

import pandas as pd
import requests
import os
```

```python
from dotenv import load_dotenv

load_dotenv()
api_key_y_string = 'YELP_API_KEY2' # replace with your Yelp API key variable name
api_key_y = os.getenv(api_key_y_string)
```

```python
for offset in range(0,150,50):
    url = f"https://api.yelp.com/v3/businesses/search"
    params = {
        'limit': 50,
        'location': 'Toronto',
        'term': 'hotel',
        'categories': 'hotel',
        'offset': offset
    }

    headers = {"accept": "application/json","Authorization": 'bearer ' + api_key_y}
    response = requests.get(url, headers=headers, params=params).json()['businesses']
    pd.json_normalize(response)
```

```python
pd.json_normalize(response)
```

```python
df = pd.json_normalize(response)
```
